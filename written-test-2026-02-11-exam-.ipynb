# 1. Load the ﬁle pollution.csv, explore the data showing size and synthetic information



```python
import pandas as pd 

df1 = pd.read_csv('written-test-2026-02-11-exam-.csv')

df1.info()
df1.describe()
```

# 2. keep only the columns ’Date’,’Time’,’CO(GT)’,’NOx(GT)’, ’PT08.S5(O3)’,’NMHC(GT)’,’NO2(GT)’,’C6H6(GT)’ then drop all the rows containing null values  

```python
df2 = df1.dropna()
df3 = df2[['Date','Time','CO(GT)','NOx(GT)', 'PT08.S5(O3)','NMHC(GT)','NO2(GT)','C6H6(GT)']]
```

```python
df3
```

# Substitute the columns ’Date’,’Time’ with a new column ’DateTime’ obtained encoding the two strings ;you can do this using the to_datetime method of pandas with an appropriate format string; 





```python
df4 = df3.copy()
df4 = df4.drop(columns=['Date','Time'])
dt = df3['Date'] + ' ' + df3['Time']
```

```python
df4['DateTime'] = pd.to_datetime(dt) 
```

```python
df4
```


# use as target ’C6H6(GT)’ and use for the prediction the other ﬁve numeric columns; the ’DateTime’; the date will not be used for the prediction, but only for the presentation of the results, as required below 

```python
X = df4.drop(columns=['C6H6(GT)','DateTime'])

y = df4['C6H6(GT)']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```


# 3. ﬁt a linear regression model on the training data, evaluate RMSE and R–square on the training and on the test data, show the values . . 3 pt


```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import GridSearchCV
from math import sqrt
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error

lr = LinearRegression()
lr_fitted = lr.fit(X_train,y_train)
y_pred_lr_train = lr.predict(X_train)
y_pred_lr_test = lr.predict(X_test)

rmse_lr_train = sqrt(mean_squared_error(y_train,y_pred_lr_train))
rmse_lr_test = sqrt(mean_squared_error(y_test,y_pred_lr_test))

r2_lr_train = r2_score(y_train,y_pred_lr_train)
r2_lr_test = r2_score(y_test,y_pred_lr_test)

print(f"Training RMSE for LR: {rmse_lr_train:.2f}")
print(f"Testing RMSE for LR: {rmse_lr_test:.2f}")
print(f"Training R2 for LR: {r2_lr_train:.2f}")
print(f"Training R2 for LR: {r2_lr_test:.2f}")
```

# 4. plot the true and the predicted target values for the test set, using for the x axis the ’Date’ column 


```python
import matplotlib.pyplot as plt
plot_df = pd.DataFrame({
    'Date': df3['Date'].loc[y_test.index],
    'y_test': y_test,
    'y_pred': y_pred_lr_test
})
plot_df=plot_df.sort_values('Date')
plt.scatter(x=plot_df['Date'],y=plot_df['y_test'])
plt.scatter(x=plot_df['Date'],y=plot_df['y_pred'])
plt.legend(labels=['y_true','y_pred'])
plt.xlabel('Date')
plt.ylabel('y_true vs. y_pred')
```

# 5. ﬁt a polynomial regression model trying several maximum degrees, up to 4, optimising the model with cross validation evaluate RMSE and R–square on the training and on the test data, show the optimal degree of the polynomial and the results . . . . . . . . . . . . . . . . . . . . . . . . . . . . . . . . 4 pt


```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV


pr = Pipeline([('poly', PolynomialFeatures()),
                  ('linear', LinearRegression(fit_intercept=False))])
grid_pr = {'poly__degree':[1,2,3,4]}
gs_pr = GridSearchCV(estimator=pr, param_grid=grid_pr, scoring='neg_root_mean_squared_error',cv=5)

gs_pr.fit(X_train,y_train)
gs_pr.best_params_
```

# 6. plot the true and the predicted target values for the test set, using for the x axis the ’Date’ column . . . . . . . . . . . . . . . . . . . . . . . . . . . . . . . . . . . 1 pt


```python
pr_best = Pipeline([('poly', PolynomialFeatures(degree=4)),
                  ('linear', LinearRegression(fit_intercept=False))])
pr_best_fitted = pr_best.fit(X_train,y_train)
y_pred_pr_best = pr_best_fitted.predict(X_test)
```

```python
plot_df_best = pd.DataFrame({
    'Date': df3['Date'].loc[y_test.index],
    'y_test': y_test,
    'y_pred': y_pred_pr_best
})
plot_df_best = plot_df_best.sort_values('Date')
plt.scatter(x=plot_df_best['Date'],y=plot_df_best['y_test'])
plt.scatter(x=plot_df_best['Date'],y=plot_df_best['y_pred'])
plt.legend(labels=['y_true','y_pred'])
plt.xlabel('Date')
plt.ylabel('y_true vs. y_pred')
```

```python
rmse_pr_test = sqrt(mean_squared_error(y_test,y_pred_pr_best))
r2_pr_test = r2_score(y_test,y_pred_pr_best)
print(f"Testing RMSE for LR: {rmse_pr_test:.2f}")
print(f"Training R2 for LR: {r2_pr_test:.2f}")
```

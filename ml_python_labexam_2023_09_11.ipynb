# 2. Eliminate totally null columns and totally null rows, eliminate columns with less than 1/3 of non null values; ﬁll the remaining NaN values with the mean of the column 


```python
pd.isnull(df1)
```

```python
#len(pd.isnull(df['X00']))
pd.isnull(df1['X00']).value_counts(df1.count)[True]
```

```python
# eliminating columns which are more than 2/3 null values
df2 = df1.copy()
for col in df1.columns:
    if pd.isnull(df1[col]).value_counts(df1.count)[True] >= 0.67:
        df2 = df2.drop(columns=[col])
```

```python
df2 = df2.dropna()
```

# 3. Drop Time, convert Date from string to datetime and group by Date using mean as aggregate  function 


```python
df3 = df2.drop(columns=['Time'])
```

```python
df3['Date'] = pd.to_datetime(df3['Date'],format='mixed')
```

```python
df3['Date']
```

```python
df = df3.groupby(['Date']).mean()
```

```python
df
```

# 4. Preparation of the boolean matrix 
• Discretise continuous values with two bins, kmeans strategy and onehot-dense encoding

• Discretization/encoding generates 0/1 values; convert the binary values obtained into boolean, as requested by Apriori


```python
from sklearn.preprocessing import KBinsDiscretizer
kb = KBinsDiscretizer(n_bins=2,encode='onehot-dense', strategy='kmeans')
transformed_data = kb.fit_transform(df)
```

```python
transformed_data
```

# 5. Set the names of two columns generated by the discretisation of each attribute A to A_low, A_high (with discretisation/one-hot-encoding, each original column generates two columns, the ﬁrst is for the low values, the second for the high values) 


```python
new_cols = pd.Index(kb.get_feature_names_out()).str.replace('_0.0','_low').str.replace('_1.0','_high')
new_cols
```

```python
transformed_data.astype(bool)
```

```python
df_ = pd.DataFrame(transformed_data.astype(bool), columns=new_cols)
```

```python
df_
```

# 6. ﬁnd a value of min_support such that the Apriori algorithm generates at least 8 frequent itemsets with at least 2 items, output the result 


```python
from mlxtend.frequent_patterns import apriori
min_sup = 0.4
target_itemset = 8
while True:
    frequent_itemsets = apriori(df_, min_support=min_sup, use_colnames=True)
    frequent_itemsets['length'] = frequent_itemsets['itemsets'].apply(len)
    n_multi = (frequent_itemsets['length']>=2).sum()
    if n_multi >= target_itemset:
        break
    min_sup-=0.01

print(f"min support: {min_sup:.2f}, number of itemsets with at least 2 items: {n_multi}")
```

# 7. ﬁnd the minimum metric threshold such that at least 100 association rules are extracted from the frequent itemsets found and show the metrics used and the best 10 rules by descending conﬁdence and support

```python
from mlxtend.frequent_patterns import association_rules
min_thres = 0.9
target_rules = 100
while True:
    rules = association_rules(frequent_itemsets, metric='confidence',min_threshold=min_thres)
    if len(rules) >= target_rules:
        break
    min_thres-=0.01
```

```python
rules
```

```python
print(f" min threshold: {min_thres:.2f}, number of rules: {len(rules)}")
```

```python
rules.sort_values(['confidence','support'],ascending=False).head(10)
```
